1. **Ma'lum bir sanadagi valyuta kurslarini olish**: '2023-12-15' sanasi uchun valyuta kurslarini olib, chop eting.

```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)
data = response.json()

for currency in data:
    if currency["Ccy"] in ["USD", "EUR", "RUB"]:
        print(f"{currency['Ccy']} = {currency['Rate']} so`m")
```

2. **USD ni UZS ga aylantirish**: '2023-12-15' sanasida USD kursini toping va 100 USD ning UZS dagi ekvivalentini hisoblang.

```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)
data = response.json()

usd_rate = None

for currency in data:
    if currency["Ccy"] == "USD":
        usd_rate = float(currency["Rate"])
        break

if usd_rate:
    amount_usd = 100
    amount_uzs = amount_usd * usd_rate
    
    print(f"USD kursi: {usd_rate} so'm")
    print(f"{amount_usd} USD = {amount_uzs:,.2f} so`m")
else:
    print("USD topilmadi")
```

3. **10,000 UZS dan yuqori kurslarga ega valyutalarni ro'yxatlash**: 10,000 UZS dan yuqori kursga ega valyutalarni chop eting.


```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    print("10 000 UZS dan yuqori kurslar:\n")
    
    for currency in data:
        rate = float(currency["Rate"])
        
        if rate > 10000:
            print(f"{currency['Ccy']} - {rate:,.2f} so`m")
else:
    print("Xatolik yuz berdi:", response.status_code)
```

4. **Eng kam valyuta kursini topish**: Eng past kursga ega valyutani aniqlang.

```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print("Eng kam valyuta kursi:\n")
    
    min_rate = float('inf')
    min_currency = ""
    
    for currency in data:
        rate = float(currency["Rate"])
        if rate < min_rate:
            min_rate = rate
            min_currency = currency["Ccy"]
    
    print(f"{min_currency} - {min_rate:,.2f} so`m")
else:
    print("Xatolik yuz berdi:", response.status_code)
```

5. **O'rtacha valyuta kursini hisoblash**: Barcha ro'yxatga olingan valyutalar orasida o'rtacha valyuta kursini hisoblang.


```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)
data = response.json() if response.status_code == 200 else []

rates = [float(x["Rate"]) for x in data]

average_rate = sum(rates) / len(rates) if rates else 0

print(f"O'rtacha valyuta kursi: {average_rate:,.2f} so'm")
```

6. **Valyuta belgisiga ko'ra kurslarni filtrlash**: EUR uchun kurslarni olib, chop eting.

```python
import requests

date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

response = requests.get(url)
data = response.json() if response.status_code == 200 else []

eur = next((x for x in data if x["Ccy"] == "EUR"), None)

if eur:
    rate = float(eur["Rate"])
    print(f"EUR kursi ({date}): {rate:,.2f} so`m")
else:
    print("EUR topilmadi")
```

9. **Eng kuchli va eng zaif valyutani aniqlash**: Kurslarga ko'ra eng kuchli va eng zaif valyutani aniqlang.

```python
import requests

# Sana
date = "2023-12-15"
url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date}/"

# So'rov
response = requests.get(url)
data = response.json() if response.status_code == 200 else []

if data:
    # Eng kuchli valyuta
    strongest = max(data, key=lambda x: float(x["Rate"]))
    # Eng zaif valyuta
    weakest = min(data, key=lambda x: float(x["Rate"]))

    print(f"Eng kuchli valyuta: {strongest['Ccy']} - {float(strongest['Rate']):,.2f} so'm")
    print(f"Eng zaif valyuta: {weakest['Ccy']} - {float(weakest['Rate']):,.2f} so'm")
else:
    print("Ma'lumot topilmadi")
```

10. **Bir qator sanalardagi kurslarni olish**: '2023-01-01' dan '2023-01-07' gacha bo'lgan kurslarni olish.

```python
import requests
from datetime import datetime, timedelta

# Sana oralig'i
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 1, 7)

# Har bir sanani aylantirish
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date_str}/"
    
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print(f"\nKurslar {date_str} sanada:")
        for currency in data:
            print(f"{currency['Ccy']} - {float(currency['Rate']):,.2f} so'm")
    else:
        print(f"{date_str} sanasi uchun ma'lumot topilmadi")
    
    current_date += timedelta(days=1)
```

11. **Bir hafta davomida valyuta kurslarini taqqoslash**: 2023 yil yanvar oyining birinchi haftasida USD kurslaridagi o'zgarishlarni chop eting.

```python
import requests
from datetime import datetime, timedelta

# Sana oralig'i: 2023-01-01 â†’ 2023-01-07
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 1, 7)

current_date = start_date
print("USD kurslari 2023 yil yanvar oyining 1-haftasida:\n")

while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    url = f"https://cbu.uz/uz/arkhiv-kursov-valyut/json/all/{date_str}/"
    
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        usd = next((x for x in data if x["Ccy"] == "USD"), None)
        if usd:
            rate = float(usd["Rate"])
            print(f"{date_str}: {rate:,.2f} so`m")
        else:
            print(f"{date_str}: USD kursi topilmadi")
    else:
        print(f"{date_str}: Ma'lumot topilmadi")
    
    current_date += timedelta(days=1)
```
