### Ishning maqsadlari

O'tilgan mavzularni mustahkamlang:

- Api bilan tanishuv

- Ma'lumotlarni ajratib olish ko'nikmasi.






#### Ishni tekshirish uchun qanday yuborish kerak
  Yakuniy fayl (.ipynb) kengaytmasida bo'lishi kerak.

---

Shu havolaga orqali ma'lumotlarga ega bo'lasiz: https://cbu.uz/uz/arkhiv-kursov-valyut/veb-masteram/

API : https://cbu.uz/uz/arkhiv-kursov-valyut/json/


1. **Ma'lum bir sanadagi valyuta kurslarini olish**: '2023-12-15' sanasi uchun valyuta kurslarini olib, chop eting.

2. **USD ni UZS ga aylantirish**: '2023-12-15' sanasida USD kursini toping va 100 USD ning UZS dagi ekvivalentini hisoblang.

3. **10,000 UZS dan yuqori kurslarga ega valyutalarni ro'yxatlash**: 10,000 UZS dan yuqori kursga ega valyutalarni chop eting.


4. **Eng kam valyuta kursini topish**: Eng past kursga ega valyutani aniqlang.

5. **O'rtacha valyuta kursini hisoblash**: Barcha ro'yxatga olingan valyutalar orasida o'rtacha valyuta kursini hisoblang.


6. **Valyuta belgisiga ko'ra kurslarni filtrlash**: EUR uchun kurslarni olib, chop eting.

7. **Ma'lum bir sanada EUR ni boshqa valyutaga aylantirish**: '2023-01-01' sanasida 50 EUR ni JPY ga aylantiring.

8. **Yaqin Sharq valyutalari uchun kurslarni chop etish**: Yaqin Sharq mamlakatlarining valyutalari uchun kurslarni filtrlang va chop eting.

9. **Eng kuchli va eng zaif valyutani aniqlash**: Kurslarga ko'ra eng kuchli va eng zaif valyutani aniqlang.

10. **Bir qator sanalardagi kurslarni olish**: '2023-01-01' dan '2023-01-07' gacha bo'lgan kurslarni olish.

11. **Bir hafta davomida valyuta kurslarini taqqoslash**: 2023 yil yanvar oyining birinchi haftasida USD kurslaridagi o'zgarishlarni chop eting.
